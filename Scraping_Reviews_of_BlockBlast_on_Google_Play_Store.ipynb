Block Blast! merupakan permainan di HP yang sedang populer dimainkan oleh pelajar sekolah masa kini. Anda dapat melihatnya dimanapun terutama di sekolah. Dengan maraknya game ini, tentunya menimbulkan sentimen mereka terhadap game tersebut baik itu berupa ulasan positif, negatif, atau netral yang dapat membantu developer mengembangkan gamenya. Maka dari itu, pada proyek analisis sentimen ini saya mengambil topik ulasan BlockBlast di Google Playstore.

Untuk dapat melakukan scrapping data, diperlukan library google-play-scrapper. Dengan library ini kita dapat mengambil data berupa ulasan aplikasi apapun yang ada di Google Playstore.

**Library yang digunakan:**

- google_play_scrapper untuk mengambil data ulasan aplikasi di Google Play Store.
- pandas untuk membaca, menyimpan, dan mengolah data ulasan dalam bentuk DataFrame.
- numpy untuk manipulasi array.

```python
from google_play_scraper import app
import pandas as pd
import numpy as np
```

```python
from google_play_scraper import Sort, reviews

result, continuation_token = reviews(
    'com.block.juggle',
    lang='id',
    country='id',
    sort=Sort.MOST_RELEVANT,
    count=20000,
    filter_score_with=None
)
```

Ulasan yang diambil dari Google Playstore terbatas hanya pada ulasan berbahasa Indonesia dan negara Indonesia. Dengan jumlah sampel yang diambil yaitu 20000 data.

```python
df_scrapped = pd.DataFrame(np.array(result),columns=['review'])

df_scrapped = df_scrapped.join(pd.DataFrame(df_scrapped.pop('review').tolist()))

df_scrapped.head()
```

```python
my_df = df_scrapped[['userName', 'score','at', 'content']]
```

Kolom data yang disimpan ke dalam file baru hanya username (pengguna yang meninggalkan ulasan), score (rating aplikasi dari pengguna), at (tanggal ulasan dibuat), dan content (isi ulasan) yang dinilai relevan untuk analisis sentimen.
